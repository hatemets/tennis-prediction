```python
seed = 35
pd.set_option('display.max_columns', 50) # display max 50 cols at once
```

```python
df = pd.read_csv(f"./data/matches_2020.csv")

relevant_cols = [
    'surface', 'tourney_date', 'winner_id', 'loser_id',
    'winner_hand', 'loser_hand', 'winner_hand', 'loser_hand',
    'winner_ht', 'loser_ht', 'winner_age', 'loser_age',
    'score', 'best_of', 'minutes', 'round', 'winner_rank', 'loser_rank',
    'winner_rank_points', 'loser_rank_points'
]

df = df[relevant_cols]
df.dropna(subset=relevant_cols, inplace=True)
df.rename(columns={'winner_id': 'player1_id', 'loser_id': 'player2_id',
                     'winner_hand': 'player1_hand', 'loser_hand': 'player2_hand',
                     'winner_ht': 'player1_ht', 'loser_ht': 'player2_ht',
                     'winner_age': 'player1_age', 'loser_age': 'player2_age',
                     'winner_rank': 'player1_rank', 'loser_rank': 'player2_rank',
                     'winner_rank_points': 'player1_rank_points', 'loser_rank_points': 'player2_rank_points'},
            inplace=True)

df['player1_win'] = 1

np.random.seed(seed)
rows_to_switch = np.random.rand(len(df)) < 0.5

# Switch columns for the selected rows
switch_columns = {
    'player1_id': 'player2_id', 'player2_id': 'player1_id',
    'player1_hand': 'player2_hand', 'player2_hand': 'player1_hand',
    'player1_ht': 'player2_ht', 'player2_ht': 'player1_ht',
    'player1_age': 'player2_age', 'player2_age': 'player1_age',
    'player1_rank': 'player2_rank', 'player2_rank': 'player1_rank',
    'player1_rank_points': 'player2_rank_points', 'player2_rank_points': 'player1_rank_points'
}

for col, switched_col in switch_columns.items():
    df.loc[rows_to_switch, [col, switched_col]] = df.loc[rows_to_switch, [switched_col, col]].values

# Invert player1_win for the switched rows
df.loc[rows_to_switch, 'player1_win'] = 0
df
```

```python
X = df.drop('player1_win', axis=1)
y = df['player1_win']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=seed)

# Preprocessing
categorical_features = ['surface', 'round']
numerical_features = ['tourney_date', 'player1_id', 'player2_id', 'player1_hand', 'player2_hand', 
                      'player1_ht', 'player2_ht', 'player1_age', 'player2_age', 'best_of', 
                      'minutes', 'player1_rank', 'player2_rank', 'player1_rank_points', 'player2_rank_points']

assert len(numerical_features) == len(set(numerical_features)), "Duplicate in numerical features"
assert len(categorical_features) == len(set(categorical_features)), "Duplicate in categorical features"


# Preprocessing
preprocessor = make_column_transformer(
    (StandardScaler(), numerical_features),
    (OneHotEncoder(), categorical_features)
)

X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)

```
